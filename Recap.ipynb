## Slicing

```python
my_list1  = [11,21,31,41,51,61,71,81]

# QUESTION : I want to print all elements after 3rd location
# Corresponding to 3rd location index is 2
# my_list1[start_index:end_index]
# if end_index is the last element of the list , I can skip
print(my_list1[2:]) 
# I want to print all elements within 2nd and 5th location 
# 2nd location is index = 1
# 5th location is index = 4
# 2nd location is actually corresponding to 21
# 5th location is actually corresponding to 51
# Python is weird --- it will print till last index - 1
print(my_list1[1:4]) 
```

```python
my_list1  = [11,21,31,41,51,61,71,81]
# QUESTION : I want to print all alternate elements after 3rd location
# my_list1[start_index:end_index:stepsize]
print(my_list1[2::3]) 


```
